6 Explain the concept of generators in python and how they are defined

---> A generator in Python is a special type of function that allows you to generate values one at a time instead of storing them all in memory.
---> Generators are useful when you have large amount of data or when you don't need all results at once.
---> You define a generator just like a normal function, but instead of return, you use the yield keyword.


```python
def squares(n):
    for i in range(1, n+1):
        yield i*i
        
s = squares(5)

for num in s:
    print(num)
```

7 What are the advantages of using a generator function over a regular function

--> A regular function returns all values at once which takes more memory.
--> A generator produces one value at a time-it does not store the entire sequence in memory.

-->Generators generate values only when needed. Regular functions calculate everything immediately and return all results. This makes generators faster and efficient for large or infinite sequences.

9 Explain the purpose and usage of map() function in python

The map() function is used to apply a function to every item of an iterable and return a map object. It allows you to process all elements without using a loop manually.

```python
def square(x):
    return x*x

numbers = [1, 2, 3, 4, 5]
squared_numbers = map(square, numbers)
print(list(squared_numbers))
```

10 What is the difference between map(), reduce(), filter() function in python

map() function : Apply a function to every item in an iterable and return a new iterable of results.
filter() function: select items from an iterable that meet a condition.
reduce() function: combine all elements of an iterable into a single value using a binary function.

```python
#map()--> map(function, iterable)
numbers = [1, 2, 3, 4, 5]
doubled = map(lambda x: x*2, numbers)
print(list(doubled))
```

```python
#filter()---> filter(function, iterable)
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))
```

```python
#reduce()---> from functools import reduce
#reduce(function, iterable)

from functools import reduce
numbers = [1, 2, 3, 4]
summation = reduce(lambda x, y: x+y, numbers)
print(summation)
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total
l1 = [1, 2, 3, 4, 5, 6]
print(sum_of_even_numbers(l1))
```

```python
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda ch: ch not in vowels, text))

str1 = "Annapurna"
result = remove_vowels(str1)
print("string without vowels:", result)
    
```
